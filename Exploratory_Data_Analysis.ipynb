# **Ship Mode:**

```python
sns.set_theme(style="darkgrid")
sns.countplot(x='Ship Mode', data=data, palette="YlOrBr")
plt.title("Ship Mode")
plt.show()
```

# **Top Sub-Categories:**

```python
subcategory = pd.DataFrame(data.groupby(['Sub-Category'])[['Quantity']].sum().sort_values('Quantity',ascending=False))
subcategory
```

```python
# Plot Bar Plot for Top Selling Sub-Category
sns.set_theme(style="darkgrid")
sns.barplot(data = subcategory, y = subcategory.index, x = subcategory.Quantity, palette = "Paired")
plt.title("Top Selling Sub-Category: ")
plt.show()
```

# **Profitable Region:**

```python
region_analysis = pd.DataFrame(data.groupby(['Region'])['Profit'].sum().reset_index())
region_analysis
```

```python
# Plot Pie Chart
explode = [0, 0, 0, 0.1]

colors = [(64/255, 64/255, 64/255, 1), (152/255, 251/255, 152/255, 1), (255/255, 127/255, 80/255, 1), (255/255, 255/255, 0/255, 1)]

plt.pie(region_analysis.Profit, labels=region_analysis.Region, startangle=90, autopct="%1.0f%%", explode=explode, shadow=True, colors=colors)
plt.title("Most Profitable by Region")

plt.show()
```

## **Regional Ship Modes:**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Region', hue='Ship Mode', data=data, palette='Blues')
plt.title("Region wise Ship Mode count")
plt.xlabel('Region')
plt.ylabel("Ship Mode count")
plt.grid()
plt.show()
```

# **Statewise Sales vs Profits:**

```python

# computing top states in terms of sales from first 10 observations
top_states_s = data.groupby("State").Sales.sum().nlargest(n=10)

# computing top states in terms of profit from first 10 observations
top_states_p = data.groupby("State").Profit.sum().nlargest(n = 10)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (8,4),fontsize = 12, color = 'coral')
top_states_p.plot(kind = 'bar',figsize = (8,4),fontsize = 12, color = 'teal')
plt.xlabel('States',fontsize = 14)
plt.ylabel('Total sales',fontsize = 14)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```
