# **K-Means Clustering**

```python
cluster_sum_of_square = np.empty((0,))
clusters_range = np.arange(1, 11)

for k in clusters_range:
    model = KMeans(n_clusters=k)
    model.fit(data)
    cluster_sum_of_square = np.append(cluster_sum_of_square, model.inertia_)
```

# **Elbow Method**

```python
plt.plot(clusters_range, cluster_sum_of_square)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Square')
plt.grid()
plt.show()
```

# **Training the Model**

```python
model = KMeans(n_clusters = 3, init ='k-means++', max_iter=200, n_init=10, random_state=11)
predictions = model.fit_predict(data)
predictions
```

```python
x = data.values
x
```

# **Cluster Visualization**

```python
plt.scatter(x[predictions==0,0], x[predictions==0,1], s=50, c='red', label='Iris-setosa')
plt.scatter(x[predictions==1,0], x[predictions==1,1], s=50, c='green', label='Iris-versicolor')
plt.scatter(x[predictions==2,0], x[predictions==2,1], s=50, c='blue', label='Iris-virginica')

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=50, c='black', label='Centroids')
plt.legend()
plt.grid()
plt.show()
```
